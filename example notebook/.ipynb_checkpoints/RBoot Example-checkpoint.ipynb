```python
import rboot
import pandas_datareader.data as web
import datetime
import numpy as np

%matplotlib inline
import matplotlib.pylab as plt
import seaborn
```

## Get some data

```python
""" Get Sample Data """
# The sample series length
N = 100

# Get sample data
spy = web.DataReader('SPY', 'yahoo', datetime.date(2000,1,1))['Adj Close']

# Take the most recent N+1 values
spy = spy[-N-1:]

# Compute Log Returns
log_rets = np.log(spy) - np.log(spy.shift(1))
log_rets = log_rets.dropna()
```

# Set Parameters

```python
bootstraps = 100
rho = 1.0 # Standard concentration of bootstraps around the mean. This value can be increased or decreased to achieve desired concentration
replacement_percentage = 1.0 # This value is the default. Decrease to reduce variability and increase to increase variability.
```

## Bootstrap the returns

```python
bs_series = rboot.replacement_bootstrap(series = log_rets, 
                                        bootstraps = bootstraps, 
                                        replacement_percentage = replacement_percentage, 
                                        rho = rho)
```

## Plot the trajectories and Bootstrap sampling distribution

```python
montecarlo = np.array([np.cumsum(v) for v in bs_series]).T
plt.plot(montecarlo);
fig,ax = plt.subplots()
seaborn.distplot(montecarlo[-1,:],ax=ax);
```
